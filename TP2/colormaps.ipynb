```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path = '/content/drive/MyDrive/Cursos_Posgrado/Procesamiento_Imagenes-UNS/TP1/mapa_temperatura.jpg'
image = cv2.imread(path)
b,g,r = cv2.split(image)
rgb_image = cv2.merge([r,g,b])
matrix_image = cv2.imread(path,0)
```

```python
#Mapas de colores existentes: https://matplotlib.org/stable/tutorials/colors/colormaps.html

fig1, axs = plt.subplots(3,3)
fig1.set_size_inches(18,18)

plt.subplot(3,3,1)
plt.imshow(rgb_image)
plt.title("Original")

plt.subplot(3,3,2)
plt.imshow(matrix_image, cmap='gray')
plt.title("Gray")

plt.subplot(3,3,3)
plt.imshow(matrix_image, cmap='binary')
plt.title("Binary")

plt.subplot(3,3,4)
plt.imshow(matrix_image, cmap='rainbow')
plt.title("Rainbow")

plt.subplot(3,3,5)
plt.imshow(matrix_image, cmap='hot')
plt.title("Hot")

plt.subplot(3,3,6)
plt.imshow(matrix_image, cmap='winter')
plt.title("Winter")

plt.subplot(3,3,7)
plt.imshow(matrix_image, cmap='prism')
plt.title("Prism")

plt.subplot(3,3,8)
plt.imshow(matrix_image, cmap='inferno')
plt.title("Inferno")

plt.subplot(3,3,9)
plt.imshow(matrix_image, cmap='terrain')
plt.title("Terrain")

fig1.tight_layout()
plt.show()

```
